```python
batch_size = config.batch_size

class AugmentedDataLoader:
    def __init__(self, dataloader, augment_fn):
        self.dataloader = dataloader
        self.augment_fn = augment_fn
        self.__iter_obj = None

    def __iter__(self):
        self.__iter_obj = iter(self.dataloader)
        return self

    def __next__(self):
        batch = next(self.__iter_obj)
        return self.augment_fn(batch)

    def __len__(self):
        return len(self.dataloader)

def augment_batch(batch):
    # Unpack the batch using the correct keys
    inputs_embeds = batch['inputs_embeds'] 
    labels = batch['labels']
    vel = batch['vel']
    
    # Create a copy for augmentation
    augmented_inputs = inputs_embeds.clone()
    
    # Apply random shifts along the time axis (vertical)
    for i in range(inputs_embeds.shape[0]):
        # Get max shift amount (limit to 25% of the time dimension)
        max_shift = max(1, int(inputs_embeds.shape[1] * 0.25))
        shift_amount = torch.randint(-max_shift, max_shift+1, (1,)).item()
        
        if shift_amount == 0:
            continue  # No shift needed
        
        # Create a temporary copy to avoid in-place modification issues
        temp = augmented_inputs[i].clone()
        
        if shift_amount > 0:  # shift down
            if shift_amount >= inputs_embeds.shape[1]:
                # Too large shift, just zero the tensor
                augmented_inputs[i, :, :] = 0
            else:
                # Use the temporary copy for the source data
                augmented_inputs[i, shift_amount:, :] = temp[:-shift_amount, :]
                augmented_inputs[i, :shift_amount, :] = 0
        else:  # shift up (shift_amount is negative)
            abs_shift = abs(shift_amount)
            if abs_shift >= inputs_embeds.shape[1]:
                # Too large shift, just zero the tensor
                augmented_inputs[i, :, :] = 0
            else:
                # Use the temporary copy for the source data
                augmented_inputs[i, :-abs_shift, :] = temp[abs_shift:, :]
                augmented_inputs[i, -abs_shift:, :] = 0
    
    # Replace the original inputs with augmented ones
    batch['inputs_embeds'] = augmented_inputs
    return batch

# Create the data loaders
train_dataloader_aug = AugmentedDataLoader(
    DataLoader(train_data, batch_size=batch_size, shuffle=True),
    augment_batch
)
test_dataloader = DataLoader(test_data, batch_size=batch_size, shuffle=True)

# Visualize some examples
for i, X in enumerate(train_dataloader_aug):
    if i == 0:
        for j in range(4):
            f, ax = plt.subplots(1, 2, figsize=(10, 5))
            ax[0].imshow(X['inputs_embeds'][j, :, :].swapaxes(0, 1) - X['inputs_embeds'][j, :, :].mean(), 
                         aspect=.1, vmin=-1, vmax=1, cmap='seismic')
            ax[0].set_title("Augmented Input")
            ax[1].step(np.pad(X['vel'][j, :], (0, 1), 'edge'), np.arange(X['vel'][j, :].shape[-1] + 1))
            ax[1].invert_yaxis()
            ax[1].set_title("Velocity")
            ax[1].grid()
            plt.tight_layout()
            plt.show()
```

### Change the prediction head

```python
class EnhancedVelpredHead(nn.Module):
    def __init__(self, config):
        super().__init__()
        # Create a more sophisticated prediction network with multiple layers
        self.dropout = nn.Dropout(0.1)
        
        # First dense layer
        self.dense1 = nn.Linear(config.hidden_size, config.hidden_size)
        self.act_fn = nn.GELU()
        self.norm = nn.LayerNorm(config.hidden_size, eps=1e-12)
        
        # Second dense layer for prediction
        self.predictions = nn.Linear(config.hidden_size, config.vel_size)
        self.predictions.decoder = Identity()
        
        # Store velocity min/max for denormalization
        self.vel_min = config.vel_min
        self.vel_max = config.vel_max
        
    def forward(self, sequence_output):
        # Take only the CLS token (first token)
        hidden_states = sequence_output[:, 0, :]
        
        # Apply intermediate layers with dropout for better generalization
        hidden_states = self.dense1(hidden_states)
        hidden_states = self.act_fn(hidden_states)
        hidden_states = self.dropout(hidden_states)
        hidden_states = self.norm(hidden_states)
        
        # Final prediction
        logits = self.predictions(hidden_states)
        
        # Denormalize the output from [-1, 1] to actual velocity range
        output = self.vel_min + (logits + 1) * (self.vel_max - self.vel_min) * 0.5
        
        return output

# Create the enhanced model head
model.cls = EnhancedVelpredHead(config)
print(model)
```

```python
def count_parameters(model):
    table = pd.DataFrame(columns=['Name', 'Parameter'])
    total_params = 0
    i = 0
    for name, parameter in model.named_parameters():
        if not parameter.requires_grad: continue
        param = parameter.numel()
        total_params+=param
        table.loc[i] = [name] + [param]
        i += 1
    display(table)
    print(f"Total Trainable Params: {total_params}")
    return total_params
    
count_parameters(model)
```

## Training

### Set up training hyperparameters

```python
# Create combined loss function
class CombinedLoss(nn.Module):
    def __init__(self, alpha=0.5, beta=0.3, gamma=0.2):
        super(CombinedLoss, self).__init__()
        self.l1_loss = nn.L1Loss()
        self.mse_loss = nn.MSELoss()
        self.alpha = alpha  # L1 loss weight
        self.beta = beta    # MSE loss weight
        self.gamma = gamma  # Gradient loss weight
        
    def forward(self, pred, target):
        # L1 loss for overall accuracy
        l1 = self.l1_loss(pred, target)
        
        # MSE loss for larger differences
        mse = self.mse_loss(pred, target)
        
        # Gradient loss to preserve velocity transitions (geological boundaries)
        # Calculate gradients using finite differences
        pred_grad = pred[:, 1:] - pred[:, :-1]
        target_grad = target[:, 1:] - target[:, :-1]
        grad_loss = self.l1_loss(pred_grad, target_grad)
        
        # Combined loss
        total_loss = self.alpha * l1 + self.beta * mse + self.gamma * grad_loss
        
        return total_loss

# Optimizer with learning rate scheduler
if config.optim == "radam":
    optim = RAdam(model.parameters(), lr=config.lr)
else:
    optim = torch.optim.Adam(model.parameters(), lr=config.lr)

# Learning rate scheduler with warmup
if config.warmup > 0:
    scheduler = torch.optim.lr_scheduler.OneCycleLR(
        optim, 
        max_lr=config.lr,
        total_steps=config.epoch * len(train_dataloader),
        div_factor=25,
        pct_start=config.warmup/config.epoch,
        anneal_strategy='cos'
    )
else:
    scheduler = None

# Use the custom combined loss
loss_fn = CombinedLoss(alpha=0.5, beta=0.3, gamma=0.2)

# Number of epochs
epochs = config.epoch
```

### Train

```python
%matplotlib notebook

plt.ion()
f, ax = plt.subplots(figsize=(4, 3))
```

```python
from transformers.modeling_outputs import BaseModelOutputWithPoolingAndCrossAttentions

# Define a custom function to update our run_velpred
def run_velpred_enhanced(
    model, optim, loss_fn, scheduler, train_dataloader, test_dataloader, vel_size, epochs, device,
    tmp_dir, patience=None, plot=False, f=None, ax=None, projection=None
):
    total_time = time.time()
    avg_train_loss = []
    avg_valid_loss = []
    time_per_epoch = []
    
    # Initialize early stopping
    if patience is not None:
        checkpoint = os.path.join(tmp_dir, str(os.getpid()) + "checkpoint.pt")
        early_stopping = EarlyStopping(patience=patience, verbose=True, path=checkpoint)
    
    # For storing best validation loss
    best_val_loss = float('inf')
    
    for epoch in range(epochs):
        epoch_time = time.time()
        model.train()
        loop_train = tqdm(train_dataloader, leave=True)
        losses_train = 0
        
        for i, batch in enumerate(loop_train):
            optim.zero_grad()
            
            # Get data and move to device
            inputs_embeds = batch['inputs_embeds'].to(device)
            labels = batch['labels'].to(device)
            
            # Handle different label dimensions
            if labels.dim() > 2:
                labels = labels[:, 0, :]
            
            # Apply projection to inputs if provided
            if projection is not None:
                inputs_embeds = projection(inputs_embeds.float())
            
            # Forward pass
            outputs = model(inputs_embeds=inputs_embeds)
            
            # Select appropriate model outputs
            if hasattr(outputs, 'logits') and outputs.logits.dim() == 3:
                pred = outputs.logits[:, 0, :]
            elif hasattr(outputs, 'logits'):
                pred = outputs.logits
            else:
                pred = outputs
            
            # Calculate loss
            loss = loss_fn(pred, labels.float())
            
            # Backward pass
            loss.backward()
            
            # Gradient clipping to prevent exploding gradients
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            
            # Update parameters
            optim.step()
            
            # Update scheduler if provided
            if scheduler is not None:
                scheduler.step()
                
            losses_train += loss.item()
            loop_train.set_description(f'Epoch {epoch}')
            loop_train.set_postfix(loss=loss.item())
        
        # Evaluate on validation set
        model.eval()
        loop_valid = tqdm(test_dataloader, leave=True)
        losses_valid = 0
        
        with torch.no_grad():
            for i, batch in enumerate(loop_valid):
                inputs_embeds = batch['inputs_embeds'].to(device)
                labels = batch['labels'].to(device)
                
                if labels.dim() > 2:
                    labels = labels[:, 0, :]
                
                if projection is not None:
                    inputs_embeds = projection(inputs_embeds.float())
                
                outputs = model(inputs_embeds=inputs_embeds)
                
                if hasattr(outputs, 'logits') and outputs.logits.dim() == 3:
                    pred = outputs.logits[:, 0, :]
                elif hasattr(outputs, 'logits'):
                    pred = outputs.logits
                else:
                    pred = outputs
                
                loss = loss_fn(pred, labels.float())
                losses_valid += loss.item()
                
                loop_valid.set_description(f'Validation {epoch}')
                loop_valid.set_postfix(loss=loss.item())
        
        # Calculate average losses
        avg_train_loss.append(losses_train / len(train_dataloader))
        avg_valid_loss.append(losses_valid / len(test_dataloader))
        
        # Print epoch statistics
        print(f"Epoch time: {time.time() - epoch_time:.2f} s")
        time_per_epoch.append(time.time() - epoch_time)
        print(f"Total time elapsed: {time.time() - total_time:.2f} s")
        print("---------------------------------------")
        
        # Plot training progress if requested
        if plot:
            ax.cla()
            ax.plot(np.arange(1, epoch+2), avg_train_loss, 'b', label='Training Loss')
            ax.plot(np.arange(1, epoch+2), avg_valid_loss, 'orange', label='Validation Loss')
            ax.legend()
            ax.set_title("Loss Curve")
            ax.set_xlabel("Epoch")
            ax.set_ylabel("Avg Loss")
            f.canvas.draw()
        
        # Save best model and check for early stopping
        if patience is not None:
            early_stopping(avg_valid_loss[-1], model)
            if early_stopping.early_stop:
                print("Early stopping triggered")
                break
    
    # Load best model when using early stopping
    if patience is not None:
        model.load_state_dict(torch.load(checkpoint))
    
    return model, avg_train_loss, avg_valid_loss, time_per_epoch

# Define the projection: mapping from the velocity input size (271) to the model's hidden size (256)
projection = nn.Linear(271, 256).to(device)

# Run the enhanced training function using the augmented data loader
model, avg_train_loss, avg_valid_loss, time_per_epoch = run_velpred_enhanced(
    model, optim, loss_fn, scheduler, train_dataloader_aug, test_dataloader, config.vel_size, 
    epochs, device, tmp_dir=config.parent_dir, patience=config.patience, 
    plot=True, f=f, ax=ax, projection=projection
)

%matplotlib inline
plt.ioff()
```

```python
# First, set up matplotlib for notebook display
%matplotlib inline
plt.rcParams['figure.figsize'] = [12, 6]  # Make figures a bit larger

# Acquisition information
x0 = 0.14  # Near offset
dx = 0.09  # Offset increment
dt = 0.008  # Time step

# Calculate evaluation metrics
def calculate_metrics(pred, true):
    # Mean Absolute Error
    mae = torch.mean(torch.abs(pred - true)).item()
    # Mean Squared Error
    mse = torch.mean((pred - true) ** 2).item()
    # R2 score
    mean_true = torch.mean(true)
    ss_tot = torch.sum((true - mean_true) ** 2)
    ss_res = torch.sum((true - pred) ** 2)
    r2 = (1 - ss_res / ss_tot).item()
    return {'MAE': mae, 'MSE': mse, 'R2': r2}

model.eval()

# Get a larger sample for better evaluation
num_samples = 8
idx = torch.randint(len(test_data), (num_samples,))
print(f"Evaluating samples with indices: {idx}")

inputs_embeds = test_data.encodings['inputs_embeds'][idx].to(device)
labels = test_data.encodings['vel'][idx].to(device)

# Track metrics for all samples
all_metrics = []

with torch.no_grad():
    # Apply the projection before passing to model
    inputs_embeds_256 = projection(inputs_embeds.float())
    sample_outputs = model(inputs_embeds=inputs_embeds_256)

# Create a figure for each sample
for i, (X, pred, true) in enumerate(zip(
    inputs_embeds.cpu(), 
    sample_outputs.cpu() if not hasattr(sample_outputs, 'logits') else sample_outputs.logits.cpu(),
    labels.cpu()
)):
    # Calculate metrics for this sample
    metrics = calculate_metrics(pred, true)
    all_metrics.append(metrics)
    
    # Create a figure with 1 row and 2 columns
    f, ax = plt.subplots(1, 2, figsize=(12, 6))
    
    # Plot the input seismic data
    im = ax[0].imshow(
        (X - X.mean()).detach().swapaxes(0, 1), 
        aspect=1.1, vmin=-1, vmax=1, cmap='seismic',
        extent=[x0, x0 + 20 * dx, 271 * dt, 0]
    )
    ax[0].set_title("Input Seismic Data", fontsize=14)
    ax[0].set_xticks([x0, 0.5, 1, 1.5, 1.9])
    ax[0].set_xticklabels([".14", "0.5", "1.0", "1.5", "1.9"])
    ax[0].set_yticks(np.arange(0, 2.5, 0.5))
    ax[0].yaxis.set_minor_locator(AutoMinorLocator(2))
    ax[0].set_ylabel("Time (s)")
    ax[0].set_xlabel("Offset (km)")
    
    # Add a colorbar
    cbar = plt.colorbar(im, ax=ax[0])
    cbar.set_label('Amplitude')
    
    # Plot the velocity profiles - true and predicted
    depth_scale = np.arange(0, 40*9, 40*9/(271))
    ax[1].step(true, depth_scale, 'k', linewidth=2, label="True Velocity")
    ax[1].step(pred, depth_scale, 'r', linewidth=2, label="Predicted")
    
    # Calculate and plot the residuals
    residual = pred - true
    ax_twin = ax[1].twiny()
    ax_twin.step(residual, depth_scale, 'g--', linewidth=1.5, label="Residual", alpha=0.7)
    ax_twin.set_xlabel("Residual (m/s)", color='g')
    ax_twin.tick_params(axis='x', colors='g')
    ax_twin.set_xlim([-500, 500])  # Set appropriate limits for residuals
    
    # Style the velocity plot
    ax[1].invert_yaxis()
    ax[1].grid(True, linestyle='--', alpha=0.7)
    ax[1].legend(loc='upper right')
    ax[1].set_xlabel("Velocity (m/s)")
    ax[1].set_ylabel("Depth (m)")
    ax[1].set_title(f"Velocity Profile - Sample {i}", fontsize=14)
    
    # Add metrics text to the plot
    metrics_text = f"MAE: {metrics['MAE']:.2f}\nMSE: {metrics['MSE']:.2f}\nR²: {metrics['R2']:.3f}"
    ax[1].text(0.05, 0.95, metrics_text, transform=ax[1].transAxes, 
              bbox=dict(facecolor='white', alpha=0.8), verticalalignment='top')
    
    plt.tight_layout()
    plt.show()

# Calculate and display average metrics
avg_metrics = {metric: np.mean([m[metric] for m in all_metrics]) for metric in all_metrics[0]}
print("\nAverage Metrics Across All Samples:")
for metric, value in avg_metrics.items():
    print(f"{metric}: {value:.4f}")
```
